```python
import pandas as pd

# Load the dataset
cars = pd.read_csv("cleaned_usedCars.csv")

# Check if the necessary columns exist
required_columns = {"Company", "BodyStyle"}
if not required_columns.issubset(cars.columns):
    missing = required_columns - set(cars.columns)
    raise KeyError(f"Missing columns in dataset: {missing}")

# Group the cars by 'company' and 'body_type'
grouped_cars = cars.groupby(["Company", "BodyStyle"])

# Iterate through each group and print details
for (company, body_type), group in grouped_cars:
    print(f"Company: {company}, Body Type: {body_type}")
    print(group.head(), "\n")  # Print the first few rows of each group
```
